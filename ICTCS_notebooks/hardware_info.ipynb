```python
import subprocess

def run_command(cmd):
    result = subprocess.run(cmd, shell=True, capture_output=True, text=True)
    print(result.stdout)

print("=== CPU Info ===")
run_command("lscpu")

print("\n=== Memory Info ===")
run_command("free -h")

print("\n=== GPU Info ===")
run_command("nvidia-smi")

print("\n=== OS Info ===")
run_command("lsb_release -a")
```
